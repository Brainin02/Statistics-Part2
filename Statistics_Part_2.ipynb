```python
# Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#interpret the results
import numpy as np
from scipy import stats


def z_test(sample_data, population_mean, population_std, sample_size):

    sample_mean = np.mean(sample_data)


    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))


    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))


    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_statistic:.2f}")
    print(f"P-value: {p_value:.4f}")


    alpha = 0.05
    if p_value < alpha:
        print(f"Reject the null hypothesis: There is a significant difference between the sample mean and the population mean.")
    else:
        print(f"Fail to reject the null hypothesis: There is no significant difference between the sample mean and the population mean.")

sample_data = [50, 52, 51, 54, 50, 49, 53, 55, 50, 51]
population_mean = 50
population_std = 2
sample_size = len(sample_data)
z_test(sample_data, population_mean, population_std, sample_size)
```

```python
# Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


def two_tailed_z_test(sample_data, population_mean, population_std):

    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)


    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    alpha = 0.05
    z_critical = stats.norm.ppf(1 - alpha / 2)

    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))


    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-statistic: {z_statistic:.2f}")
    print(f"P-value: {p_value:.4f}")


    if p_value < alpha:
        print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
    else:
        print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")


    plot_decision_region(z_statistic, z_critical, alpha)


def plot_decision_region(z_statistic, z_critical, alpha):

    z_values = np.linspace(-4, 4, 1000)
    normal_pdf = stats.norm.pdf(z_values)


    plt.figure(figsize=(10, 6))
    plt.plot(z_values, normal_pdf, label='Standard Normal Distribution', color='b')


    plt.fill_between(z_values, normal_pdf, where=(z_values <= -z_critical), color='red', alpha=0.5)
    plt.fill_between(z_values, normal_pdf, where=(z_values >= z_critical), color='red', alpha=0.5)


    plt.axvline(x=z_statistic, color='g', linestyle='--', label=f"Z-statistic = {z_statistic:.2f}")


    plt.axvline(x=-z_critical, color='r', linestyle='--', label=f"Critical Z = {-z_critical:.2f}")
    plt.axvline(x=z_critical, color='r', linestyle='--', label=f"Critical Z = {z_critical:.2f}")

    plt.title(f"Two-Tailed Z-test (Œ± = 0.05)")
    plt.xlabel("Z-value")
    plt.ylabel("Probability Density")
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.show()


sample_data = np.random.normal(loc=52, scale=8, size=30)
population_mean = 50
population_std = 8
two_tailed_z_test(sample_data, population_mean, population_std)
```

```python
# Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_null, mu_alt, sigma, n, alpha=0.05, test_type='right'):
    """
    Visualize Type I and Type II errors for a one-sample Z-test.

    Parameters:
    - mu_null: Mean under the null hypothesis (H0)
    - mu_alt: Mean under the alternative hypothesis (H1)
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level (default = 0.05)
    - test_type: 'left', 'right', or 'two-tailed'
    """

    se = sigma / np.sqrt(n)

    if test_type == 'right':
        z_crit = norm.ppf(1 - alpha)
        critical_value = mu_null + z_crit * se
    elif test_type == 'left':
        z_crit = norm.ppf(alpha)
        critical_value = mu_null + z_crit * se
    elif test_type == 'two-tailed':
        z_crit = norm.ppf(1 - alpha / 2)
        critical_value_right = mu_null + z_crit * se
        critical_value_left = mu_null - z_crit * se
    else:
        raise ValueError("test_type must be 'left', 'right', or 'two-tailed'")


    x = np.linspace(mu_null - 4*se, mu_alt + 4*se, 1000)


    null_pdf = norm.pdf(x, mu_null, se)
    alt_pdf = norm.pdf(x, mu_alt, se)


    plt.figure(figsize=(10, 6))
    plt.plot(x, null_pdf, label='Null Hypothesis (H‚ÇÄ)', color='blue')
    plt.plot(x, alt_pdf, label='Alternative Hypothesis (H‚ÇÅ)', color='green')


    if test_type == 'right':
        plt.fill_between(x, 0, null_pdf, where=(x >= critical_value), color='red', alpha=0.3, label='Type I Error (Œ±)')
        plt.axvline(critical_value, color='red', linestyle='--')

        plt.fill_between(x, 0, alt_pdf, where=(x < critical_value), color='purple', alpha=0.3, label='Type II Error (Œ≤)')
    elif test_type == 'left':
        plt.fill_between(x, 0, null_pdf, where=(x <= critical_value), color='red', alpha=0.3, label='Type I Error (Œ±)')
        plt.axvline(critical_value, color='red', linestyle='--')
        plt.fill_between(x, 0, alt_pdf, where=(x > critical_value), color='purple', alpha=0.3, label='Type II Error (Œ≤)')
    elif test_type == 'two-tailed':
        plt.axvline(critical_value_right, color='red', linestyle='--')
        plt.axvline(critical_value_left, color='red', linestyle='--')
        plt.fill_between(x, 0, null_pdf, where=(x >= critical_value_right) | (x <= critical_value_left), color='red', alpha=0.3, label='Type I Error (Œ±)')
        plt.fill_between(x, 0, alt_pdf, where=(x < critical_value_right) & (x > critical_value_left), color='purple', alpha=0.3, label='Type II Error (Œ≤)')

    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()


visualize_type1_type2_errors(
    mu_null=100,
    mu_alt=105,
    sigma=15,
    n=30,
    alpha=0.05,
    test_type='right'
)
```

```python
# Write a Python program to perform an independent T-test and interpret the results.
import numpy as np
from scipy import stats

def independent_t_test(group1, group2, equal_var=True):
    """
    Perform an independent two-sample t-test and interpret the results.

    Parameters:
    - group1: list or array of numeric values (sample 1)
    - group2: list or array of numeric values (sample 2)
    - equal_var: boolean, True for standard t-test, False for Welch's t-test
    """

    t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=equal_var)


    print(f"Group 1 mean: {np.mean(group1):.2f}")
    print(f"Group 2 mean: {np.mean(group2):.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_value:.4f}")


    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: The group means are significantly different.")
    else:
        print("Fail to reject the null hypothesis: No significant difference between the group means.")


np.random.seed(42)
group1 = np.random.normal(loc=75, scale=10, size=30)
group2 = np.random.normal(loc=80, scale=10, size=30)

independent_t_test(group1, group2, equal_var=True)
```

```python
 #  Perform a paired sample T-test using Python and visualize the comparison results.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def paired_t_test(before, after):
    """
    Perform a paired sample t-test and visualize the results.

    Parameters:
    - before: scores or measurements before treatment
    - after: scores or measurements after treatment
    """
    # Perform paired t-test
    t_stat, p_value = stats.ttest_rel(before, after)

    # Summary statistics
    mean_before = np.mean(before)
    mean_after = np.mean(after)

    # Print results
    print(f"Mean Before: {mean_before:.2f}")
    print(f"Mean After:  {mean_after:.2f}")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_value:.4f}")

    # Interpret result
    alpha = 0.05
    if p_value < alpha:
        print("‚úÖ Reject the null hypothesis: There is a significant difference.")
    else:
        print("‚ùå Fail to reject the null hypothesis: No significant difference.")

    # Visualize the comparison
    plt.figure(figsize=(8, 5))
    plt.plot([1]*len(before), before, 'bo', label='Before')
    plt.plot([2]*len(after), after, 'go', label='After')
    for i in range(len(before)):
        plt.plot([1, 2], [before[i], after[i]], 'k-', alpha=0.3)
    plt.xticks([1, 2], ['Before', 'After'])
    plt.ylabel('Score')
    plt.title('Paired Sample T-test: Before vs After')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Simulated data: scores before and after a training program
np.random.seed(42)
before_scores = np.random.normal(loc=70, scale=8, size=25)
after_scores = before_scores + np.random.normal(loc=5, scale=5, size=25)

# Run the test
paired_t_test(before_scores, after_scores)
```

```python
#  Simulate data and perform both Z-test and T-test, then compare the results using Python.
import numpy as np
from scipy import stats

def simulate_and_compare_tests(mu_population=100, sigma_population=15, sample_size=30, alpha=0.05):
    # Simulate data
    np.random.seed(42)
    sample = np.random.normal(loc=mu_population + 3, scale=sigma_population, size=sample_size)  # slight shift

    # Known population mean and std deviation
    mu_0 = mu_population
    sigma = sigma_population
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)

    print(f"\nSample Mean: {sample_mean:.2f}")
    print(f"Sample Std Dev: {sample_std:.2f}")
    print(f"Population Mean (Œº‚ÇÄ): {mu_0}")
    print(f"Sample Size: {sample_size}\n")

    # Z-test (known population std deviation)
    z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(sample_size))
    p_z = 2 * (1 - stats.norm.cdf(abs(z_stat)))

    # T-test (unknown population std deviation)
    t_stat, p_t = stats.ttest_1samp(sample, popmean=mu_0)

    # Results
    print("Z-test results (population std known):")
    print(f"Z-statistic: {z_stat:.3f}")
    print(f"P-value: {p_z:.4f}")
    print("=>", "Reject H‚ÇÄ" if p_z < alpha else "Fail to reject H‚ÇÄ")

    print("\nT-test results (population std estimated):")
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_t:.4f}")
    print("=>", "Reject H‚ÇÄ" if p_t < alpha else "Fail to reject H‚ÇÄ")

# Run the comparison
simulate_and_compare_tests()
```

```python
# Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data: list or numpy array of sample observations
    - confidence: confidence level (default = 0.95)

    Returns:
    - (lower_bound, upper_bound): tuple representing the confidence interval
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t-critical value

    margin_of_error = t_crit * std_err
    ci_lower = mean - margin_of_error
    ci_upper = mean + margin_of_error

    print(f"Sample Mean: {mean:.2f}")
    print(f"Confidence Level: {confidence*100:.0f}%")
    print(f"Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")

    return (ci_lower, ci_upper)

# Example usage with simulated data
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=25)
confidence_interval(sample_data, confidence=0.95)
```

```python
# Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy import stats

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Parameters:
    - data: list or numpy array of sample observations
    - confidence: confidence level (default = 0.95)

    Returns:
    - margin_of_error: the calculated margin of error
    """
    n = len(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_crit = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t-critical value

    margin_of_error = t_crit * std_err
    print(f"Sample Size: {n}")
    print(f"Standard Error: {std_err:.4f}")
    print(f"T-critical value (df={n-1}): {t_crit:.4f}")
    print(f"Margin of Error at {int(confidence * 100)}% confidence: ¬±{margin_of_error:.2f}")
    return margin_of_error

# Example usage with simulated sample data
np.random.seed(1)
sample_data = np.random.normal(loc=70, scale=12, size=40)
calculate_margin_of_error(sample_data, confidence=0.95)
```

```python
# Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
def bayes_theorem(prior, sensitivity, specificity):
    """
    Apply Bayes' Theorem to a diagnostic test scenario.

    Parameters:
    - prior: prior probability of having the disease (e.g., 0.01)
    - sensitivity: P(Test+ | Disease) ‚Äî true positive rate
    - specificity: P(Test- | No Disease) ‚Äî true negative rate

    Returns:
    - posterior: P(Disease | Test+)
    """
    # Prior probabilities
    P_D = prior
    P_not_D = 1 - P_D

    # Likelihoods
    P_pos_given_D = sensitivity
    P_pos_given_not_D = 1 - specificity

    # Marginal probability of a positive test
    P_pos = (P_pos_given_D * P_D) + (P_pos_given_not_D * P_not_D)

    # Posterior using Bayes' Theorem
    P_D_given_pos = (P_pos_given_D * P_D) / P_pos

    print(f"Posterior Probability P(Disease | Test+): {P_D_given_pos:.4f}")
    return P_D_given_pos

# Example usage
prior_prob = 0.01         # 1% disease prevalence
sensitivity = 0.99        # True positive rate
specificity = 0.95        # True negative rate

bayes_theorem(prior_prob, sensitivity, specificity)
```

```python
# Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
#data
import pandas as pd
import numpy as np

def calculate_expected_frequencies(observed_data):
    """
    Calculate the expected frequencies for a Chi-square test given observed data.

    Parameters:
    - observed_data: Pandas DataFrame of observed frequencies

    Returns:
    - expected_df: DataFrame of expected frequencies
    """
    # Convert to DataFrame if not already
    if not isinstance(observed_data, pd.DataFrame):
        observed_data = pd.DataFrame(observed_data)

    # Calculate row totals, column totals, and grand total
    row_totals = observed_data.sum(axis=1)
    col_totals = observed_data.sum(axis=0)
    grand_total = observed_data.values.sum()

    # Compute expected frequencies
    expected = np.outer(row_totals, col_totals) / grand_total
    expected_df = pd.DataFrame(expected, index=observed_data.index, columns=observed_data.columns)

    print("Observed Frequencies:\n", observed_data)
    print("\nExpected Frequencies:\n", expected_df.round(2))

    return expected_df

# Example observed data
observed = pd.DataFrame({
    "Tea":    [30, 20],
    "Coffee": [20, 30]
}, index=["Male", "Female"])

# Calculate expected frequencies
calculate_expected_frequencies(observed)
```

```python
#  Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution .
import numpy as np
from scipy.stats import chisquare

def chi_square_goodness_of_fit(observed, expected=None):
    """
    Perform a Chi-square goodness-of-fit test.

    Parameters:
    - observed: list or array of observed frequencies
    - expected: list or array of expected frequencies (optional)

    Returns:
    - Chi-square statistic and p-value
    """
    observed = np.array(observed)

    if expected is None:
        # If expected is not given, assume uniform distribution
        expected = np.full_like(observed, fill_value=observed.sum() / len(observed))

    chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

    print("Observed:", observed)
    print("Expected:", expected)
    print(f"\nChi-square statistic: {chi2_stat:.3f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < 0.05:
        print("Reject the null hypothesis: The data does not fit the expected distribution.")
    else:
        print(" Fail to reject the null hypothesis: The data fits the expected distribution.")

    return chi2_stat, p_value

# Example: Test if a die is fair
observed_rolls = [8, 9, 10, 11, 12, 10]
chi_square_goodness_of_fit(observed_rolls)
```

```python
#  Implement an F-test using Python to compare the variances of two random samples.
import numpy as np
from scipy.stats import f

def f_test_variance(sample1, sample2):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
    - sample1: list or numpy array of data for sample 1
    - sample2: list or numpy array of data for sample 2

    Returns:
    - F-statistic and p-value
    """

    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    n1 = len(sample1)
    n2 = len(sample2)

    f_stat = var1 / var2 if var1 > var2 else var2 / var1

    df1 = n1 - 1
    df2 = n2 - 1


    p_value = 1 - f.cdf(f_stat, df1, df2)

    print(f"Variance of Sample 1: {var1:.3f}")
    print(f"Variance of Sample 2: {var2:.3f}")
    print(f"F-statistic: {f_stat:.3f}")
    print(f"P-value: {p_value:.4f}")

    alpha = 0.05
    if p_value < alpha:
        print("\n Reject the null hypothesis: The variances are significantly different.")
    else:
        print("\n Fail to reject the null hypothesis: The variances are not significantly different.")

    return f_stat, p_value
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=55, scale=15, size=30)

f_test_variance(sample1, sample2)
```

```python
# Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, levene, probplot

def check_anova_assumptions(groups):
    """
    Check the assumptions for ANOVA: normality, independence, and equal variance.

    Parameters:
    - groups: list of groups (each group should be a numpy array or list of data)

    Returns:
    - None
    """
    # Check normality assumption using Shapiro-Wilk test for each group
    print("Checking normality using Shapiro-Wilk test:")
    for i, group in enumerate(groups):
        stat, p_value = shapiro(group)
        print(f"Group {i+1}: Shapiro-Wilk statistic = {stat:.4f}, p-value = {p_value:.4f}")
        if p_value < 0.05:
            print(f"  Group {i+1} does not follow normal distribution (p < 0.05)")
        else:
            print(f"   Group {i+1} follows normal distribution (p >= 0.05)")

    # Plot Q-Q plots to visually assess normality
    plt.figure(figsize=(12, 8))
    for i, group in enumerate(groups):
        plt.subplot(1, len(groups), i+1)
        probplot(group, dist="norm", plot=plt)
        plt.title(f"Group {i+1} Q-Q Plot")
    plt.tight_layout()
    plt.show()

    # Check equality of variances using Levene's test
    print("\nChecking equality of variances using Levene's test:")
    stat, p_value = levene(*groups)
    print(f"Levene's statistic = {stat:.4f}, p-value = {p_value:.4f}")
    if p_value < 0.05:
        print("  Variances are not equal (p < 0.05)")
    else:
        print("  Variances are equal (p >= 0.05)")

# Example usage with data for three groups
np.random.seed(42)
group1 = np.random.normal(loc=75, scale=10, size=30)  # Study Method 1
group2 = np.random.normal(loc=80, scale=12, size=30)  # Study Method 2
group3 = np.random.normal(loc=70, scale=8, size=30)   # Study Method 3

# Check ANOVA assumptions
check_anova_assumptions([group1, group2, group3])
```

```python
 #  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.
 import numpy as np
import scipy.stats as stats

# Step 1: Simulate random data from a normal distribution
np.random.seed(42)

# Simulate sample data with a true variance of 25 (standard deviation = 5)
sample_size = 30
true_variance = 25
sample_data = np.random.normal(loc=0, scale=5, size=sample_size)

# Step 2: Perform the Chi-square test for population variance
sample_variance = np.var(sample_data, ddof=1)  # sample variance (unbiased)
hypothesized_variance = 25  # hypothesized population variance

# Calculate the Chi-square test statistic
chi_square_stat = (sample_size - 1) * sample_variance / hypothesized_variance

# Step 3: Find the p-value using the Chi-square distribution
degrees_of_freedom = sample_size - 1
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, degrees_of_freedom),
                  1 - stats.chi2.cdf(chi_square_stat, degrees_of_freedom))

# Display the results
print(f"Chi-square statistic: {chi_square_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Interpretation of the results
alpha = 0.05  # significance level

if p_value < alpha:
    print("\nReject the null hypothesis: The population variance is different from the hypothesized value.")
else:
    print("\nFail to reject the null hypothesis: The population variance is not significantly different from the hypothesized value.")
```

```python
# Write a Python script to perform a Z-test for comparing proportions between two datasets or groups.
import numpy as np
import scipy.stats as stats

# Step 1: Simulate Data for Two Groups (Successes and Sample Sizes)
# Group 1: 100 successes out of 200 trials
successes1 = 100
n1 = 200

# Group 2: 120 successes out of 250 trials
successes2 = 120
n2 = 250

# Step 2: Calculate the sample proportions
p1 = successes1 / n1
p2 = successes2 / n2

# Step 3: Calculate the pooled proportion
p = (successes1 + successes2) / (n1 + n2)

# Step 4: Calculate the Z-statistic
z = (p1 - p2) / np.sqrt(p * (1 - p) * (1/n1 + 1/n2))

# Step 5: Calculate the p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

# Step 6: Display the results
print(f"Z-statistic: {z:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 7: Interpretation of results
alpha = 0.05  # Significance level

if p_value < alpha:
    print("\nReject the null hypothesis: The proportions are significantly different.")
else:
    print("\nFail to reject the null hypothesis: The proportions are not significantly different.")
```

```python
# Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Function to perform the F-test
def f_test(data1, data2, alpha=0.05):
    # Calculate variances
    var1 = np.var(data1, ddof=1)
    var2 = np.var(data2, ddof=1)

    # Calculate the F-statistic (larger variance / smaller variance)
    F = var1 / var2 if var1 >= var2 else var2 / var1
    df1 = len(data1) - 1  # degrees of freedom for data1
    df2 = len(data2) - 1  # degrees of freedom for data2

    # Calculate the critical value from the F-distribution
    critical_value = stats.f.ppf(1 - alpha, df1, df2)

    # Perform the hypothesis test
    if F > critical_value:
        result = 'Reject the null hypothesis (variances are different)'
    else:
        result = 'Fail to reject the null hypothesis (variances are equal)'

    return F, critical_value, result

# Visualize the F-distribution and the F-statistic
def plot_f_distribution(df1, df2, F, critical_value):
    # Create an F-distribution plot
    x = np.linspace(0, 5, 1000)
    y = stats.f.pdf(x, df1, df2)

    plt.plot(x, y, label='F-distribution', color='blue')
    plt.axvline(x=critical_value, color='red', linestyle='--', label='Critical Value')
    plt.axvline(x=F, color='green', linestyle='--', label='Calculated F-statistic')

    plt.title('F-distribution and F-test Result')
    plt.xlabel('F value')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example datasets
data1 = np.random.normal(loc=0, scale=1, size=100)  # Dataset 1
data2 = np.random.normal(loc=0, scale=2, size=100)  # Dataset 2

# Perform the F-test
F, critical_value, result = f_test(data1, data2, alpha=0.05)

# Output the result
print(f'F-statistic: {F}')
print(f'Critical Value: {critical_value}')
print(result)

# Plot the F-distribution and test results
plot_f_distribution(len(data1) - 1, len(data2) - 1, F, critical_value)
```
